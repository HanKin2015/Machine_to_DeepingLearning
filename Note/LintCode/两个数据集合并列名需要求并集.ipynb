# 需求：三个数据集合并列名需要求并集

```python
import pandas as pd

df1 = pd.DataFrame({'a':1, 'b':2, 'c':3, 'd':4}, index=[0])
df2 = pd.DataFrame({'a':1, 'b':2, 'e':5, 'f':6}, index=[0])
df3 = pd.DataFrame({'a':1, 'c':3, 'e':5, 'h':7}, index=[0])
df1, df2, df3
```

```python
df = pd.concat([df1, df2, df3], ignore_index=True, sort=False)
df
```

```python
df = df.fillna(0).astype(int)
df
```
